```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
from matplotlib.patches import FancyArrowPatch
import warnings
warnings.filterwarnings('ignore')

matplotlib.rcParams.update({
    'font.family': 'serif',
    'font.size': 9,
    'axes.titlesize': 10,
    'axes.labelsize': 9,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'legend.fontsize': 8,
    'figure.dpi': 150,
})
```

```python
# ── DATA ──────────────────────────────────────────────────────────────────────
# NSSF purchase index: 2019 = 100 (baseline), data through 2022
# 2020: +58.2%  → 158.2  (NSSF Retailer Survey)
# 2021: continued elevated (60% of retailers still reporting increased Black traffic)
#        Conservative estimate: ~175
# 2022: post-Bruen normalisation begins; peak enthusiasm; estimate ~168
# Natural decline back toward baseline expected as the 2020 urgency fades
nssf_years = [2019, 2020, 2021, 2022]
nssf_index = [100,  158.2, 175,  162]

# GSS personal Black gun ownership rate (VPC 2022 report — approximate midpoints)
# Note: not available 2019–2021 due to COVID survey disruption
gss_years = [1980, 1984, 1987, 1990, 1993, 1996, 1998, 2000,
              2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018]
gss_pct   = [14.8, 17.2, 16.1, 20.4, 14.5, 16.0, 17.2, 20.0,
              18.1, 14.3, 18.0, 16.6, 14.4, 14.5, 17.4, 15.4, 16.1]

# Catalyst events (within the NSSF data window — 2019–2022)
catalysts = [
    (2020.21, 'COVID-19\npandemic'),
    (2020.40, 'George Floyd\nmurdered'),
    (2021.32, 'Bruen cert\ngranted'),
    (2022.48, 'Bruen decided\n(shall-issue)'),
]

# Massachusetts legislative sequence (2023–2024) — shown separately
ma_events = [
    (2023.78, 'HD 4607\n(original bill,\nno carve-outs)'),
    (2024.06, 'S.2572: LEO\ncarve-outs\n(ψ_m Stage 1)'),
    (2024.57, 'Ch. 135\nsigned\n(grandfather)'),
    (2024.75, 'Emergency\npreamble\n(Oct 2)'),
]
```

```python
# ── FIGURE ────────────────────────────────────────────────────────────────────
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import warnings
warnings.filterwarnings('ignore')

matplotlib.rcParams.update({
    'font.family': 'serif',
    'font.size': 9,
    'axes.titlesize': 10,
    'axes.labelsize': 9,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'legend.fontsize': 8,
    'figure.dpi': 150,
})

color_nssf    = '#c0392b'   # red — purchase index
color_gss     = '#2c3e50'   # dark navy — historical ownership rate
color_cat     = '#e67e22'   # orange — catalyst events
color_bruen   = '#2980b9'   # blue — Bruen legal events
color_ma      = '#8e44ad'   # purple — MA suppression sequence
color_neutral = '#95a5a6'   # grey — grid / separators

fig = plt.figure(figsize=(9, 6.5))
gs = fig.add_gridspec(3, 1, height_ratios=[3.5, 0.08, 1.4], hspace=0)

ax_top  = fig.add_subplot(gs[0])   # NSSF purchase index + catalysts
ax_div  = fig.add_subplot(gs[1])   # thin divider
ax_bot  = fig.add_subplot(gs[2])   # MA legislative timeline

# ── TOP PANEL ─────────────────────────────────────────────────────────────────
# GSS rate on right twin axis
ax_gss = ax_top.twinx()
ax_gss.plot(gss_years, gss_pct, color=color_gss, linewidth=1.2,
            linestyle='--', marker='o', markersize=3, alpha=0.65)
ax_gss.set_ylabel('GSS personal ownership %
(historical, right axis)',
                  fontsize=7.5, color=color_gss, labelpad=4)
ax_gss.set_ylim(5, 38)
ax_gss.tick_params(axis='y', labelcolor=color_gss, labelsize=7)
ax_gss.text(2018.8, 16.5, '16.1 %\n(2018, last GSS)', fontsize=6.5,
            color=color_gss, va='center', ha='left', style='italic')

# NSSF purchase index
ax_top.fill_between(nssf_years, 100, nssf_index, alpha=0.12, color=color_nssf)
ax_top.plot(nssf_years, nssf_index, color=color_nssf, linewidth=2.4,
            marker='o', markersize=6, zorder=4)

# Baseline
ax_top.axhline(100, color=color_neutral, linewidth=0.7, linestyle=':', zorder=0)
ax_top.text(2018.82, 101.5, '2019 baseline', fontsize=6.5, color=color_neutral)

# Annotate the peak
ax_top.annotate('+58.2 % vs. 2019\n(NSSF 2020)',
                xy=(2020, 158.2), xytext=(2020.3, 166),
                fontsize=7, color=color_nssf,
                arrowprops=dict(arrowstyle='->', color=color_nssf, lw=0.8))

# Natural decline note
ax_top.annotate('Post-peak\nnormalisation',
                xy=(2022, 162), xytext=(2021.55, 145),
                fontsize=6.5, color=color_neutral,
                arrowprops=dict(arrowstyle='->', color=color_neutral, lw=0.7))

# Catalyst event verticals
for i, (ex, elabel) in enumerate(catalysts):
    ec = color_bruen if 'Bruen' in elabel else color_cat
    ax_top.axvline(ex, color=ec, linewidth=0.9, linestyle='--', alpha=0.55, zorder=1)
    # Stagger label height alternately
    yval = 188 if i % 2 == 0 else 179
    ax_top.text(ex + 0.03, yval, elabel, fontsize=6.5, color=ec,
                va='top', linespacing=1.15)

ax_top.set_xlim(2018.6, 2023.0)
ax_top.set_ylim(85, 205)
ax_top.set_xticklabels([])
ax_top.set_ylabel('NSSF Black firearm purchase index\n(2019 = 100)', fontsize=8)
ax_top.set_title(
    'Black Firearm Acquisition 2019–2022: Three Catalysts and Natural Normalisation',
    fontsize=10, fontweight='bold', pad=7
)

handles = [
    mlines.Line2D([], [], color=color_nssf, linewidth=2, marker='o', markersize=5,
                  label='NSSF purchase index (2019 = 100)'),
    mlines.Line2D([], [], color=color_gss, linewidth=1.2, linestyle='--', marker='o',
                  markersize=3, label='GSS personal ownership % (right axis, 1980–2018)'),
    mlines.Line2D([], [], color=color_cat,   linewidth=1.2, linestyle='--', label='Catalyst event'),
    mlines.Line2D([], [], color=color_bruen, linewidth=1.2, linestyle='--', label='Bruen (legal event)'),
]
ax_top.legend(handles=handles, loc='upper left', fontsize=7, framealpha=0.88, borderpad=0.5)

# ── DIVIDER ───────────────────────────────────────────────────────────────────
ax_div.set_xlim(0, 1); ax_div.set_ylim(0, 1)
ax_div.axhline(0.5, color='#bdc3c7', linewidth=1.2)
ax_div.set_xticks([]); ax_div.set_yticks([])
ax_div.set_xlabel('')
ax_div.spines[:].set_visible(False)
ax_div.text(0.01, 0.5, 'Massachusetts legislative response (2023–2024)',
            fontsize=8, color=color_ma, fontweight='bold', va='center')

# ── BOTTOM PANEL — MA TIMELINE ────────────────────────────────────────────────
# Map MA events to a 0–1 x-axis (purely chronological spacing)
n = len(ma_events)
xs = np.linspace(0.12, 0.88, n)

ax_bot.set_xlim(0, 1); ax_bot.set_ylim(0, 1)
ax_bot.axhline(0.5, color=color_ma, linewidth=1.0, alpha=0.4)
ax_bot.set_xticks([]); ax_bot.set_yticks([])
ax_bot.spines[:].set_visible(False)

for i, ((year, label), xpos) in enumerate(zip(ma_events, xs)):
    ymarker = 0.5
    ax_bot.plot(xpos, ymarker, marker='v', color=color_ma, markersize=9, zorder=4)
    # Alternate above / below
    if i % 2 == 0:
        ax_bot.text(xpos, ymarker + 0.12, label, fontsize=6.8, color=color_ma,
                    ha='center', va='bottom', linespacing=1.15)
        ax_bot.text(xpos, ymarker - 0.05, str(year)[:4], fontsize=6.2,
                    color=color_ma, ha='center', va='top', style='italic')
    else:
        ax_bot.text(xpos, ymarker - 0.12, label, fontsize=6.8, color=color_ma,
                    ha='center', va='top', linespacing=1.15)
        ax_bot.text(xpos, ymarker + 0.07, str(year)[:4], fontsize=6.2,
                    color=color_ma, ha='center', va='bottom', style='italic')

ax_bot.set_xlabel(
    'ψ_m sequence: HD 4607 → LEO carve-outs → Chapter 135 signed → Emergency preamble',
    fontsize=7.5, color=color_ma, labelpad=4
)

# Data note
fig.text(
    0.01, 0.005,
    'Top panel sources: NSSF Retailer Survey (2020, 2021); VPC/GSS (1973–2018); Buggs et al. 2022 (NIH PMC8697522). '
    'Bottom panel: Massachusetts Legislature public record.',
    fontsize=5.8, color='#7f8c8d', va='bottom'
)

plt.savefig('../figures/eq_black_gun_ownership_mt.png',
            dpi=200, bbox_inches='tight', facecolor='white')
plt.savefig('../figures/eq_black_gun_ownership_mt.pdf',
            bbox_inches='tight', facecolor='white')
plt.show()
print('Saved → figures/eq_black_gun_ownership_mt.{png,pdf}')
```
